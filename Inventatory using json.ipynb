```python
import json
import time

# Importing Inventory data from Record.json file
fd=open("records.json","r")
js=fd.read()
fd.close()
# Converting String data to Dictionary
records=json.loads(js)
#initial records converted into Json using json.dumps
# records={12345:{"name":"kitkat"         ,"rate":15,"quantity":465},
#          23456:{"name":"notebook"       ,"rate":43,"quantity":1070},
#          34567:{"name":"cookies"        ,"rate":12,"quantity":876},
#          45678:{"name":"chips"          ,"rate":10,"quantity":1445},
#          56789:{"name":"cheese"         ,"rate":55,"quantity":225},
#          67890:{"name":"bread"          ,"rate":18,"quantity":397},
#          78901:{"name":"little-hearts"  ,"rate":10,"quantity":586}
# }

# Displaying Menu
print("Product"," " ,"Rate")
print("-"*50)
for i in records.keys():
    print(records[i]["name"],":-",records[i]["rate"])
print("-"*50)

# Taking Inputs from the user about their details and purchase    
c_name=str(input("Enter your name      :"))
c_ph  =str(input("Enter your phone.NO  :"))

bill=[]
while True:
    id_pro=str(input())
    if id_pro in records:
        qn_pro=int(input())
        # If we're having equal or more quantity then the user wants
        if (records[id_pro]["quantity"]>=qn_pro):
                tran=id_pro+"     "+records[id_pro]["name"]+"     "+str(qn_pro)+"     "+str(records[id_pro]["rate"])+"     "+str(records[id_pro]["rate"]*qn_pro)
                bill.append(tran)
                # Updating Inventory in Dictionary
                records[id_pro]["quantity"]-=qn_pro
        
        # If we're less quantity then the user wants
        else:
            print("We only have ",records[id_pro]["quantity"],records[id_pro]["quantity"])
            print("Would you like to Purchase it")
            ch=str(input(print("Y/N")))
            
            print("-"*80)
            
            # If user wants to purchase the whole quantity for that product
            if ch=="y" or ch=="Y":
                tran1=id_pro+"     "+records[id_pro]["name"]+"     "+str(records[id_pro]["quantity"])+"     "+str(records[id_pro]["rate"])+"     "+str(records[id_pro]["rate"]*records[id_pro]["quantity"])
                bill.append(tran1)
                # Updating Inventory in Dictionary
                records[id_pro]["quantity"]=0
            else:
                break
                   
    else:
        break
    
print("-"*80)
print("                                   BILL                                            \n")
print("                                 SHOP NAME                                            ")
print("-"*80)
print("Product ID","  ","Product Name","  ","Product Quantity","  ","Product Rate(Rs)","  ","Product Price(Rs)")
print("-"*80)
for i in bill:
    print(i)
    # Generating CSV Transection Detail
    sale=c_name+" "*7+c_ph+" "*7+i+time.ctime()+"\n"
    
    # Adding Transection on Sales File
    fd=open("sales.txt","a")
    fd.write(sale)
    fd.close
    
    
# Converting Inventory Dictionary to String
js=json.dumps(records)

# Updating Inventory and Saving in to my Records.json
fd=open("records.json","w")
fd.write(js)
fd.close()


print('')
print("---------------------------------------------")
print("  Thanks for your order, Inventory Updated!  ")
print("---------------------------------------------")
```
